```python
import pandas as pd
import numpy as np
pd.options.display.max_rows = 100
pd.options.display.max_columns= 150

df = pd.read_csv('LoanStats_2018Q4.csv',skiprows=1);
print(df.shape);
df.head()
```

```python
df['int_rate'];
```

```python
def int_rate_to_float(mystring):
  if isinstance(mystring,str):
    return(float(mystring.strip().strip('%')))
  else: return mystring
```

```python
def clean_title(title):
  if isinstance(title,str):
    return title.strip().title()
  else:
    return "Unknown"
```

```python
df = df.drop(['url','member_id','desc','id'], axis=1) #Drop empty columns
df = df.drop(index=[128413,128412]) #Drop last two rows because they're only NaN
```

```python
df['int_rate'] = df['int_rate'].apply(int_rate_to_float) #Convert Int_Rate to floats
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title) #Clean the titles up
```

```python
df['term'];
```

```python
df['term'][0]
```

```python
def turn_to_int(inputStr):
  return int(inputStr.strip().strip(' months'))
```

```python
print(turn_to_int(df['term'][0]))
```

```python
df['term'] = df['term'].apply(turn_to_int)
```

```python
df['loan_status'][:10] #See what the data looks like
```

```python
t = []

for x in df['loan_status']: #iterate through the df column 'loan_status"
  if x == "Current" or x == "Fully Paid": #check if the string says "Current" or "Fully Paid"
    t.append(1) #If yes, append one to the list t
  else:
    t.append(0) #Else, append zero to the list t

df['loan_status_is_great'] = pd.Series(t) #Set the column 'loan_status_is_great' to a series created from the list t
```

```python
df['loan_status_is_great'][:10] #Make sure it worked right
```

```python
df['last_pymnt_d'][:10] #See what the data looks like
```

```python
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d']) #convert to datetime
```

```python
df['last_pymnt_d'][:10]
```

```python
t0 = []
t1 = []
for x in df['last_pymnt_d']: t0.append(x.month)
for x in df['last_pymnt_d']: t1.append(x.year)
df['last_pymnt_d_month'] = pd.Series(t0)
df['last_pymnt_d_year'] = pd.Series(t1)
```
